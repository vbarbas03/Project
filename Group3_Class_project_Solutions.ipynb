## **Task 4: Anomaly, and Outlier Detection**
### **Objective**: Identify and manage anomalies in the dataset.
Actions:
- Visualize distributions of key numerical variables using boxplots or histograms to identify outliers.
- Use the Interquartile Range (IQR) method to detect outliers in at least two numerical columns (e.g., Sleep Duration, Daily Steps).
  - Calculate the Quartiles first quartile (Q1) and third quartile Q3
  - Compute the $IQR = Q3 - Q1$
  - Determine the Bounds
    - Lower Bound: $Q1 - 1.5 \times IQR$
    - Upper Bound: $Q3 + 1.5 \times IQR$
    - Identify Outliers
- Decide whether to remove or adjust outliers based on your analysis.

- Give your observations.

```python
data['age'].quantile(0.25)
```

```python
column = 'age'
```

```python
data[column].quantile(0.25)
```

```python
def iqr_outlier(data,column):
  q1 = data[column].quantile(0.25)
  q3 = data[column].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
  return outliers
```

```python
iqr_outlier(data,'age')
```

### **MinMax Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
a = [1,4,2]
min(a)
```

```python
def minmax(my_list):
  scaled_list = []
  min_a = min(my_list)
  max_b = max(my_list)
  for i in my_list:
    scaled_i = (i - min_a)/(max_b - min_a)
    scaled_list.append(scaled_i)
  return scaled_list, min_a, max_b
```

```python
minmax(a)
```

```python
scaler = MinMaxScaler()
```

### **Z-Score Normalization**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler_z = StandardScaler()
data['age_z_scaled'] = scaler_z.fit_transform(data[['age']].values.reshape(-1, 1))
```

```python
# 2. Data visualization
plt.figure(figsize=(6,4))
plt.hist(data= data, x = 'age_z_scaled', edgecolor = 'black', color = 'green', bins = 15)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

**StandardScaler** standardizes features by removing the mean and scaling to unit variance. It transforms the data to have a mean of 0 and a standard deviation of 1.

**MinMaxScaler** scales features to a fixed range, typically between 0 and 1, by subtracting the minimum value and dividing by the range (max - min)

---
**StandardScaler** preserves the shape of the original distribution, only changing its scale.

**MinMaxScaler** can alter the shape of the original distribution, especially if outliers are present
